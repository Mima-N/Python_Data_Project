# How are in-demand skills trending for Data Analysts?

### Methodology

1. Aggregate skill counts monthly
2. Plot the monthly skill count
3. Re-analyze based on percentage of total jobs
4. Plot the montly skill demand

```python
# Importing Libraries
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data cleaning
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA_US = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()
```

```python
# Create a new month column to aggregate this on a monthly basis 
df_DA_US['job_posted_month_no'] = df_DA_US['job_posted_date'].dt.month
```

```python
df_DA_US_explode = df_DA_US.explode('job_skills')
```

```python
df_DA_US_pivot = df_DA_US_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)

# Creating a new row of total to be able to sort the columns 
df_DA_US_pivot.loc['Total'] = df_DA_US_pivot.sum()

df_DA_US_pivot = df_DA_US_pivot[df_DA_US_pivot.loc['Total'].sort_values(ascending=False).index]

df_DA_US_pivot = df_DA_US_pivot.drop('Total')

df_DA_US_pivot
```

```python
# To get total number of jobs posted in each individual months
DA_totals = df_DA_US.groupby('job_posted_month_no').size()
```

```python
df_DA_US_percent = df_DA_US_pivot.div(DA_totals, axis=0) * 100

df_DA_US_percent
```

```python
# changes the month number to month names
df_DA_US_percent = df_DA_US_percent.reset_index()
df_DA_US_percent['job_posted_month'] = df_DA_US_percent['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_DA_US_percent = df_DA_US_percent.set_index('job_posted_month')
df_DA_US_percent = df_DA_US_percent.drop(columns='job_posted_month_no')

df_DA_US_percent
```

```python
df_plot = df_DA_US_percent.iloc[:, :5]
sns.lineplot(data=df_plot, dashes=False, palette='tab10', legend=False)
sns.set_theme(style='ticks')
sns.despine()
plt.title('Trending Top Skills of Data Analysts in the US')
plt.xlabel('2025')
plt.ylabel('Likelihood in Job Posting')
plt.yticks([10, 20, 30, 40, 50, 60])
plt.ylim(10, 60)

from matplotlib.ticker import PercentFormatter
ax = plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))

# Manual positioning for overlapping labels
label_positions = {
    0: 0,      # sql - use actual value
    1: 0,      # excel - use actual value
    2: 1.5,    # python - offset up
    3: -1.5,   # tableau - offset down
    4: 0       # sas - use actual value
}

for i in range(5):
    plt.text(11.2, df_plot.iloc[-1, i] + label_positions[i], df_plot.columns[i])

plt.tight_layout()
plt.show()
```

```python
label_offsets = {
    'tableau': -1.8,
    'python': 0.0,  # Python stays at its final point
}

df_plot = df_DA_US_percent.iloc[:, :5]

sns.lineplot(data=df_plot, dashes=False, palette='tab10', legend=False)
sns.set_theme(style='ticks')
sns.despine()

plt.title('Trending Top Skills of Data Analysts in the US')
plt.xlabel('2025')
plt.ylabel('Likelihood in Job Posting')
plt.yticks([10, 20, 30, 40, 50, 60])
plt.ylim(10, 60)

from matplotlib.ticker import PercentFormatter
ax = plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))

# Apply labels with offsets
for i in range(5):
    skill_name = df_plot.columns[i]
    y_position = df_plot.iloc[-1, i] + label_offsets.get(skill_name, 0)
    plt.text(11.2, y_position, skill_name)

plt.tight_layout()
plt.show()
```
